```python
class quickFindUF:
    
    id = []
    
    def __init__(self, n=None, arr=None):
        
        #If there is the array is not created
        if arr == None:
            #We construct an array
            self.id = self.__construct_array(n)
        else:
            self.id = arr
    
    def __construct_array(self, n):
        
        arr = []
        
        #Set every element of array size n to its own group
        for i in range(n):
            arr.append(i)
        
        return arr
    
    #Now we compare the value of 2 sets to see if they are in the same group

    def connected(self, i1, i2):
        return self.id[i1] == self.id[i2]
    
    #merge 2 sets if they are not in the same group
    #i1 and other elements in the same group as i1 will merge into the group of i2
    def merge(self, i1, i2):
        g1 = self.id[i1]
        g2 = self.id[i2]
        
        if self.connected(i1, i2) == False:
            for i in range(len(self.id)):
                if self.id[i] == g1:
                    self.id[i] = g2
                    
                 
```

```python
class weightedQuickUnionCompression:
    
    id = []
    weight = []
    
    def __init__(self, n=None, arr=None):
        
        self.weight = [1]*n
        #If there is the array is not created
        if arr == None:
            #We construct an array
            self.id = self.__construct_array(n)
        else:
            self.id = arr
    
    def __construct_array(self, n):
        
        arr = []
        
        #Set every element of array size n to its own group
        for i in range(n):
            arr.append(i)
        
        return arr
    
    
    def connected(self, i1, i2):
                
        #Compare the leader of two numbers
        #Set the succesor of i1 or leader of i1
        leader1 = self.findLeader(i1)
        self.id[i1] = leader1
        
        leader2 = self.findLeader(i2)
        self.id[i2] = leader2
        
        return leader1 == leader2
    
    def findLeader(self, i):
        while self.id[i] != i:
            #Compression
            i = self.id[i]
        return i
            
                    
    
    def merge(self, i1, i2):
        
        #Merge the leader of i1 to i2
        
        if self.connected(i1, i2) == False:
            
            leader1 = self.id[1]
            leader2 = self.id[2]
            
            #If weight of i1 > i2, then leader of i2 wil connect to i1
            if self.weight[leader1] > self.weight[leader2]:
        
                self.id[leader2] = i1
                self.weight[leader1] += self.weight[leader2]
                
                
            
            #If weight of i1 <= i2, then leadear of i1 will connect to i2
            else:
            
                self.id[leader1] = i2
                self.weight[leader2] += self.weight[leader1]
```

```python
import timeit
from timeit import Timer

def test(method,size):
    nums = method(size)
    
    p, q = random.randrange(size), random.randrange(size)
    
    nums.connected(p,q)
    nums.merge(p,q)
    

expSize = 100

results = []

for i in range(1000, 100000, 1000):
    
    time1 = Timer('test(quickFindUF, %d)'%i,'from __main__ import test, quickFindUF')
    qf = time1.timeit(number = expSize)

    time2 = Timer('test(weightedQuickUnionCompression, %d)'%i,'from __main__ import test, weightedQuickUnionCompression')
    qu = time2.timeit(number = expSize)
    
    results.append((i, qf, qu))


```
